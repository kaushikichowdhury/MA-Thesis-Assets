***CMU Dictionary Testing***

*To address previous inconsistencies in the syllable detection, an additional verification step was implemented using the **CMU** pronunciation dictionary accessed via NLTK.*

```python
import nltk
nltk.download('cmudict')

from nltk.corpus import cmudict
d = cmudict.dict()
```

```python
# Function to get syllable count
def syllable_count_word(word):
    word = word.lower()

    if word in d:
        return max([sum(1 for phoneme in pron if str.isdigit(phoneme[-1])) for pron in d[word]])

    return 1  # Fallback if word not in dictionary
```

```python
# Cleaning up the text for each line
import re

def clean_line(line):
    line = line.lower()
    line = re.sub(r"[^a-z\s]", "", line)
    return line
```

```python
def syllables_in_line(line):

    line = clean_line(line)
    words = line.split()

    return sum(syllable_count_word(w) for w in words)
```

```python
lines = poem.split("\n")

for line in lines:
    if line.strip() == "":
        continue

    print(syllables_in_line(line))
```

>*Now we find an accurate syllable count per line, and CMU Dictionary is also known to be able to find stress patterns. But Prosodic still remains superior in metre recognition at a close range.*
>
>*Additionally, we must also find a way to find if poems are more rhyming or in free verse.*

```python
# Function to get rhyme sound
def rhyme_part(word):
    word = word.lower()

    if word not in d:
        return None

    pronunciation = d[word][0]

    for i in range(len(pronunciation)-1, -1, -1):
        if pronunciation[i][-1].isdigit():
            return tuple(pronunciation[i:])
```

```python
import re

lines = poem.split("\n")

last_words = []

for line in lines:
    words = re.findall(r"[a-zA-Z']+", line.lower())
    if words:
        last_words.append(words[-1])

print(last_words)
```

```python
# Checking if the above words rhyme
for i in range(len(last_words)):
    for j in range(i+1, len(last_words)):
        if rhyme_part(last_words[i]) == rhyme_part(last_words[j]):
            print(f"Line {i+1} rhymes with Line {j+1}")
```

```python
# Rhyme Scheme (in case of free verse there will be no repetitive patterns of letters)
scheme = []
rhyme_map = {}
label = "ABCDEFGHIJKLMNOPQRSTUVWXYZ"
count = 0

for word in last_words:
    rp = rhyme_part(word)

    if rp not in rhyme_map:
        rhyme_map[rp] = label[count]
        count += 1

    scheme.append(rhyme_map[rp])

print("Rhyme scheme:", "".join(scheme))
```

*Thus, CMU Dictionary satisfactorily finds if a poem is  in free verse. But what if a poem does rhyme? How would that look?*
